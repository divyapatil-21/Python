## Write A Python Program To Check Duplication Of Student Identity In A Student List 


```python

student_ids = [101, 102, 103, 104, 102, 105, 101]

unique_ids = set()

duplicates = []

for i in student_ids:
    if i in unique_ids:
        duplicates.append(i)
    else:
        unique_ids.add(i)

if duplicates:
    print("Duplicate Student IDs found:", set(duplicates))
else:
    print("No duplicate student IDs found.")
```

## Write A Python Program To Generate Even Numbers From 0 To 20 And Odd Numbers From 0 To 20, Then Add Both Generated Numbers And Display The Total Result 

```python
even_numbers = [i for i in range(0, 21) if i % 2 == 0]

odd_numbers = [i for i in range(0, 21) if i % 2 != 0]

total = sum(even_numbers) + sum(odd_numbers)

print("Even numbers:", even_numbers)
print("Odd numbers:", odd_numbers)
print("Total of all even and odd numbers:", total)
```

## Write A Python Program To Get Age Of 10 Students From User And Store In A List. Display Ages That Are Greater Than 14 And Less Than 20 

```python
ages = []

for i in range(10):
    age = int(input(f"Enter age of student {i+1}: "))
    ages.append(age)

filtered_ages = [a for a in ages if a > 14 and a < 20]

print("\nAll ages:", ages)
print("Ages greater than 14 and less than 20:", filtered_ages)
```
