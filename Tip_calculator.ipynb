## Tip Calculator

```python
bill_amount = input("Enter bill amount:")
tip_amount= input("Enter tip amount:")

adult = int(input("Enter number of adult customer :"))
child=int( input("Enter number of child customer :"))

total_amount = int(bill_amount )+ int(tip_amount)
print("Total Amount:" ,total_amount)

adult_pay = total_amount/(adult+(child/2))
print(adult_pay)

child_pay= adult_pay/2

print(child_pay)
```
