```python
# Hangman 

import random
import string
from wonderwords import RandomWord

guess_list = []


r = RandomWord()
random_word = r.word().lower()   # Generating random word
print("Random Word : ", random_word)

life = 5    # chances

progress = ['_'] * len(random_word)   # progress tracker ('_' by default)
print(' '.join(progress))             # displays underscore with spaces
    
    

while life > 0:
    
    print("Life Remaining : ", life)                        # remaining lifes
    guess = input("Enter your Word Guess : ").lower()
    
    if len(guess) == 1 and guess.isalpha() :                # 1 char at a time and it should be an alphabet only

        if guess in guess_list:                             # duplicacy char check
            print("You already guessed that letter!")
            print()
            continue
            
        guess_list.append(guess)                            # appending char in guess list
        
        if guess in random_word:
            print("Right Guess")
            
            for i in range(len(random_word)):
                if random_word[i] == guess:    # matching char with random word's index
                     progress[i] = guess       # assigning char to that position of progress 
            print()
            print(' '.join(progress))          # writing progress
            print()
            
            if '_' not in progress:
                print("Bingo!!! You've Won the Game.")
                break
            
        else:                                 # if guess not in random word, life -1 
            life -= 1
            print("Wrong guess!")  
            print(' '.join(progress))
            print()
            
            if life == 0:    
                print("Better Luck Next Time")
                print('The Word was', random_word)

    else:
        print("Enter single Character at a time, it should be an Alphabet.") 
        print()
```
