```python
import pandas as pd

data = {
    "Name": ['Ravi', 'Pooja', 'Ankit', 'Neha'],
    "Marks": [78, 85, 90, 72],
    "City": ['Bhopal', 'Indore', 'Bhopal', 'Indore']
}

df = pd.DataFrame(data)
print(df[df['Marks']<80])
print(df[(df['City']== 'Indore') & (df['Marks']>70)])
print(df[df['Marks']>85][['Name', 'City']])
print(df[df['Marks'].between(70,80)])
print(df[df['City'] != 'Bhopal'])
```

```python
import pandas as pd

data = {
    "Name": ['Ravi', 'Pooja', 'Ankit', 'Neha'],
    "Marks": [78, 85, 90, 72],
    "City": ['Bhopal', 'Indore', 'Bhopal', 'Indore']
}

df = pd.DataFrame(data)
print(df.sort_values(by='Marks'))
print(df.sort_values(by= 'Marks', ascending= False))
print(df.sort_values(by='City'))
print(df.sort_values(by=['City', 'Marks'], [[True, False]] ))
```
