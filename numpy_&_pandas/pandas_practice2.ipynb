```python
import pandas as pd

data = {
    'City': ['Indore', 'Indore', 'Bhopal', 'Bhopal', 'Indore'],
    'Marks': [80, 85, 70, 90, 95],
    'Age': [20, 22, 21, 19, 20]
}
df = pd.DataFrame(data)
z=df.groupby('City').agg({'Marks': ['sum', 'mean'], 'Age': 'mean'})

y= df.groupby('City')['Marks'].mean().sort_values(ascending=False)

x=df.groupby('City', as_index=False)['Age'].mean()

v= df.groupby('City').filter(lambda x: x['Marks'].mean() > 80)


print(z)
print(y)
print(x)
print(v)
```

```python
import pandas as pd

data = {
    'Name': ['Riya', 'Priya', 'Divya'],
    'JoinDate': ['2024-01-10', '2024-03-05', '2024-02-18'],
    'Sales': [20000, 35000, 18000]
}
df = pd.DataFrame(data)
df['JoinDate']= pd.to_datetime(df['JoinDate'])

df['Year'] = df['JoinDate'].dt.year
df['Month'] = df['JoinDate'].dt.month
df['Day'] = df['JoinDate'].dt.day
df['Weekday'] = df['JoinDate'].dt.day_name()


print(df)
```
