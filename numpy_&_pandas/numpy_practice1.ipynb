## Array shapes & reshaping

```python
a = np.arange(12)
b = a.reshape(3,4)
print(a)
print(b)
print(a.shape, b.shape)
# flatten
c = b.ravel()
print(c)
```

```python
a= np.arange(24)
print(a)
b=a.reshape(4,6)
print(b)


d=b.reshape(2,12)
print(d)
```

```python
import pandas as pd

s = pd.Series([10, 20, 30], index=["A", "B", "C"])
print(s)
```

```python
import pandas as pd
sales = pd.Series([2000, 3500, 1800, 5000], index=["Mon", "Tue", "Wed", "Thu"])
print("Original Series:\n", sales)

print("Sales > 3000:\n", sales[sales > 3000])

print("Sum =", sales.sum())
print("Sorted:\n", sales.sort_values())
```
