```python
from pymongo import MongoClient
from pprint import pprint

client = MongoClient("mongodb://127.0.0.1:27017")   # already working on your machine
db = client['my_app_db']                           # choose/create a database
col = db['users']                                  # choose/create a collection

# CREATE (insert one & many)
user1 = {"name": "Divya", "role": "engineer", "age": 24}
user2 = {"name": "Amit", "role": "tester", "age": 28}
res1 = col.insert_one(user1)
res2 = col.insert_many([user2, {"name":"Riya","role":"intern","age":21}])
print("Inserted IDs:", res1.inserted_id, res2.inserted_ids)

# READ (find)
print("All users:")
for doc in col.find():
    pprint(doc)

# READ (query + projection + sort + limit)
for doc in col.find({"age": {"$gte": 25}}, {"name":1, "age":1, "_id":0}).sort("age", -1).limit(5):
    pprint(doc)

# UPDATE
col.update_one({"name":"Amit"}, {"$set": {"role":"senior tester", "age":29}})
col.update_many({"role":"intern"}, {"$inc": {"age": 1}})

# DELETE
col.delete_one({"name":"Riya"})
col.delete_many({"age": {"$lt": 22}})

# Index example (for faster queries)
col.create_index([("name", 1)], unique=False)

# Close connection when done (optional — client is lazy and can be reused)
client.close()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "http://quotes.toscrape.com"
output_csv = r"D:\python\scraped_quotes.csv"   # FIXED ✔️

all_rows = []

page = 1
while True:
    r = requests.get(f"{URL}/page/{page}/")
    if r.status_code != 200:
        break
    soup = BeautifulSoup(r.text, "lxml")

    quotes = soup.find_all("div", class_="quote")
    if not quotes:
        break

    for q in quotes:
        text = q.find("span", class_="text").get_text(strip=True)
        author = q.find("small", class_="author").get_text(strip=True)
        tags = [t.get_text(strip=True) for t in q.find_all("a", class_="tag")]
        
        all_rows.append({
            "text": text,
            "author": author,
            "tags": ",".join(tags)
        })

    page += 1

df = pd.DataFrame(all_rows)
df.to_csv(output_csv, index=False, encoding="utf-8")   # ✔️ no error now
print(f"Saved {len(df)} rows to {output_csv}")
```
