You are going to write two functions called encrypt() and decrypt() to implement the Caesar Cipher 
encryption and decryption. 
The Caesar Cipher is one of the simplest and oldest encryption techniques. It works by shifting each 
letter in the text by a fixed number of positions down or up the alphabet. 
Steps to Encrypt: 
1. Take a plain text message and a shift number (for example, 3). 
2. For each letter in the message: - Find its position in the alphabet (A=0, B=1, ... Z=25). - Add the shift number to the position to get a new position. - Replace the letter with the new shifted letter. 
3. If you go past 'Z', wrap around to the beginning of the alphabet. 
Steps to Decrypt: 
1. Take the encrypted message and the same shift number used to encrypt it. 
2. For each letter in the encrypted text: - Find its position in the alphabet. - Subtract the shift number to get the original position. - Replace it with the corresponding letter. 
3. Again, wrap around if you go before 'A'. 
Example: 
Plain text: "HELLO" 
Shift: 3 
Encryption process: 
H → K 
E → H 
L → O 
L → O 
O → R 
Encrypted message: "KHOOR"

```python
def encrypt(text, shift):
    alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l',
                'm','n','o','p','q','r','s','t','u','v','w','x','y','z']
    encrypted_text = ""

    for i in text:
        if i in alphabet:
            position = alphabet.index(i)
            new_position = (position + shift) % 26   
            encrypted_text += alphabet[new_position]
        else:
            encrypted_text += i
    return encrypted_text


def decrypt(text, shift):
    alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l',
                'm','n','o','p','q','r','s','t','u','v','w','x','y','z']
    decrypted_text = ""

    for i in text:
        if i in alphabet:
            position = alphabet.index(i)
            new_position = (position - shift) % 26   
            decrypted_text += alphabet[new_position]
        else:
            decrypted_text += i
    return decrypted_text


text = input("Enter text: ").lower()
shift = int(input("Enter how many numbers you want to shift: "))

encrypted_message = encrypt(text, shift)
print("Encrypted text:", encrypted_message)

decrypted_message = decrypt(encrypted_message, shift)
print("Decrypted text:", decrypted_message)

```
