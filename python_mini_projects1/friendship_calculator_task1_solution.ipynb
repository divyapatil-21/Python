## Friendship Calculator

You are going to write a function called calculate_friendship_score() that tests the friendship 
compatibility between two people.

1. To work out the friendship score between two names: 
2. Take both people's names and check for the number of times the letters in the word FRIEND 
occur. 
3. Then check for the number of times the letters in the word BOND occur. 
4. Combine these two totals to make a 2-digit number, which represents the Friendship Score. 
Example 
name1 = "Shahrukh Khan" 
name2 = "Salman Khan" 
Step 1: Count letters from “FRIEND” 
F occurs 0 times 
R occurs 1 time 
I occurs 1 time 
E occurs 0 times 
N occurs 2 times 
D occurs 0 times 
Total = 4 
Step 2: Count letters from “BOND” 
B occurs 0 times 
O occurs 0 times 
N occurs 2 times 
D occurs 0 times 
Total = 2 
Friendship Score = 42 

```python
def friendship_score():
    name1= input("Enter first name").lower()
    name2= input("Enter second name").lower()
    names = name1+name2
   
    
    
    word1="friend" 
    word2="bond"
    
    word_count1 = 0
    word_count2 = 0
       
    
    for i in word1 : 
        if i in names:
            word_count1 += names.count(i)
    for i in word2 : 
        if i in names:
            word_count2 +=names.count(i)
    
    combine= str(word_count1)+str(word_count2)
    print("frienship score:",combine)
friendship_score()
```
